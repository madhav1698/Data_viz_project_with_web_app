```python
import pandas as pd
from datetime import datetime as dt
import matplotlib.pyplot as plt
from pytz import utc
data=pd.read_csv("reviews.csv",parse_dates=["Timestamp"]) #parses timetsamp column as datetime
```

```python
data.head()
```

```python
data["Day"]=data["Timestamp"].dt.date
day_average= data.groupby(["Day"]).mean()
day_average
```

```python
plt.figure(figsize=(25,3))
plt.plot(day_average.index,day_average["Rating"])
```

```python
###Rating avg by week
```

```python
data["Week"]=data["Timestamp"].dt.strftime("%Y-%W") #string from time
week_avg=data.groupby(["Week"]).mean()
plt.figure(figsize=(25,3))
plt.plot(week_avg.index,week_avg["Rating"])
```

```python
##avg rating by month
data["Month"]=data["Timestamp"].dt.strftime("%Y-%m") #string from time
month_avg=data.groupby(["Month"]).mean()
plt.figure(figsize=(25,3))
plt.plot(month_avg.index,month_avg["Rating"])
month_avg
```

```python
### Average rting by month by course
```

```python
data["Month"]= data["Timestamp"].dt.strftime("%y-%m")
month_avg_by_course=data.groupby(["Month","Course Name"]).mean().unstack() #for better structure
month_avg_by_course.plot(figsize=(25,10))
```

```python
###What day are people happiest?
```

```python
data["Weekday"]=data["Timestamp"].dt.strftime("%A")
data["Daynumber"]=data["Timestamp"].dt.strftime("%w")

weekday_avg= data.groupby(["Weekday","Daynumber"]).mean()
weekday_avg=weekday_avg.sort_values("Daynumber")
weekday_avg.index.get_level_values(0) #extracts stuff at 0 index

plt.figure(figsize=(15,3))
plt.plot(weekday_avg.index.get_level_values(0),weekday_avg["Rating"])
```
